```python
import matplotlib.pyplot as plt
from pymedit import Mesh, P1Function, trunc, mmg2d
import numpy as np

import pandas as pd
```

```python
Th = Mesh('meshes/Th.mesh')
Th.plot(boundary='all')
```

```python
Tseq = P1Function(Th, "output/Tseq.sol")
Tseq.plot()
```

```python
T = P1Function(Th, "output/T.sol")
T.plot()
```

```python
TFullParallel = P1Function(Th, "output/TFullParallel.sol")
TFullParallel.plot()
```

```python
TF = P1Function(Th, "output/TF.sol")
TF.plot()
```

```python
error = P1Function(Th, np.abs(T.sol - TF.sol))
error.plot()
```

```python
error = P1Function(Th, np.abs(TFullParallel.sol - TF.sol))
error.plot()
```
