# Convergence.py 

```python
epsilon =  np.array([5e-3,7.5e-3,1e-2,2e-2,3e-2,4e-2,5e-2,6e-2,7e-2,8e-2,9e-2])
conv = np.array([
                    0.0067498212867541,
                    0.007224626860822178,
                    0.00797379783216832,
                    0.0142715059987254,
                    0.01979756278411348,
                    0.02484124263077974,
                    0.02926382820485506,
                    0.03335464893519158,
                    0.0374600448769613,
                    0.0410500739911336,
                    0.0445050405195148
])
plt.plot(np.log10(epsilon), np.log10(conv),'-o')
```

```python
i = 2
X = np.log10(epsilon[i:]).reshape(-1, 1) #epsilon[2:].reshape(-1, 1) #np.log10(epsilon[2:]).reshape(-1, 1)
y = np.log10(conv[i:]) #conv[2:] #np.log10(conv[2:])
reg = LinearRegression().fit(X, y)

b = reg.intercept_
a = reg.coef_[0] #order
a
```

```python
plt.plot(X, y,'-o')
plt.plot(X, a*X + b)
```

```python
plt.plot(10.**X, 10.**y,'-o')
plt.plot(10.**X, 10.**(a*X + b))
```
