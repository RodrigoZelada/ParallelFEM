```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# FEM

```python
eps = pd.DataFrame({'eps':10.**np.arange(0, -9, -1)})
xf = pd.read_csv("output/resitCG.txt", names = ["number of iterations"], dtype=int)
yf = pd.read_csv("output/resCG.txt", names = ["residual"])
zf = pd.read_csv("output/ERD.txt", names = ["energy error"])
wf = pd.read_csv("output/ERDeps.txt", names = ["jump error"])
df = pd.concat([eps, xf, yf, zf, wf], axis=1)
df
```

```python
print(df.to_latex())
```
