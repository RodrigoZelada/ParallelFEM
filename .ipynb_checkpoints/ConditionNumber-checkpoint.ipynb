```python
import matplotlib.pyplot as plt
from pymedit import Mesh, P1Function, trunc
import numpy as np
import pandas as pd
```

# Rectilinear case

```python
epsilon = pd.DataFrame({'eps':10.**np.arange(0, -8, -1)})
k = pd.DataFrame({'eps':np.arange(0, 8, 1)})
condD = pd.read_csv("output/condD.txt", names = ["conditioning"])
condN = pd.read_csv("output/condN.txt", names = ["conditioning"])
dfD = pd.concat([k, epsilon, condD], axis=1)
dfN = pd.concat([k, epsilon, condN], axis=1)
dfD
```

```python
# epsilon = 10.**np.arange(-8, 10, 2)
# condD = [ 2.19844e+11, 2.19847e+09, 2.20154e+07, 6.0268e+06,    6.03144e+06, 6.03871e+06, 6.03889e+06, 6.03889e+06, 6.03889e+06]
# condN = [  6.02689e+06, 6.02689e+06,  6.02684e+06, 6.02624e+06, 6.03077e+06, 6.03803e+06, 6.03822e+06, 6.03822e+06,  6.03822e+06]


plt.plot(k.values, np.log10(condD.values),'-o')
#plt.plot(np.log10(epsilon), np.log10(condN),'-o')
plt.legend(['FEM approx.'])
plt.xlabel('$k$ ($\epsilon = 10^{-k}$)')
plt.ylabel('$\log_{10}$ condition number')

plt.savefig("output/ConditionNumberSinusCase.png")

plt.plot(k.values, np.log10(condD.values),'-o')
plt.plot(k.values, np.log10(condN.values),'-o')
plt.legend(['FEM', 'Nitsche'])
plt.xlabel('$k$ ($\epsilon = 10^{-k}$)')
plt.ylabel('$\log_{10}$ condition number')

plt.savefig("output/ConditionNumberNitscheSinusCase.png")
```
